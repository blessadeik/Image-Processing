### Counting Objects

```python
# Parsing an Argument
import argparse
import imutils
import cv2
import imutils


## arparse was commented out because it can only be run on the terminal. to run on terminal, uncomment the next 4 line
## and comment out the code on line 16

# ap = argparse.ArgumentParser()
# ap.add_argument("-i", "--jar", required=True, help="path to input image")
# args = vars(ap.parse_args())
# jar = cv2.imread(args["jar"])

jarImage = cv2.imread("jar.png")

cv2.imshow("JarImage", jarImage)
cv2.waitKey(0)

# Converting our the image to grayscale
grayJar = cv2.cvtColor(jarImage, cv2.COLOR_BGR2GRAY)
cv2.imshow("Grayscale jarImage", grayJar)
cv2.waitKey(0)

# this code is to apply edge detection so we can find the outlines of objects in grayJar
outlinedImage = cv2.Canny(grayJar, 30, 150)
cv2.imshow("Outlined GrayJar", outlinedImage)
cv2.waitKey(0)

# Threshold the image by setting all pixel values less than 225 to 255 (white; foreground) and all pixel values
# >= 225 to 255 (black; background), thereby segmenting the image
threshedJar = cv2.threshold(grayJar, 225, 255, cv2.THRESH_BINARY_INV)[1]
cv2.imshow("Thresholded grayJar", threshedJar)
cv2.waitKey(0)

# This code is to find contours-outlines of the foreground objects in the thresholded image
cnts = cv2.findContours(threshedJar.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
cnts = imutils.grab_contours(cnts)
result = jarImage.copy()

# This will loop over the contours
# draw each contour on the output image with a 4px thick pink outline, then display the output contours one at a time
for c in cnts:

    cv2.drawContours(result, [c], -1, (255,192,203), 3)
    cv2.imshow("Contours", result)
    cv2.waitKey(0)
    
# draw the total number of contours found in pink
text = "There are {} objects in the image!".format(len(cnts))
cv2.putText(result, text, (10, 25),  cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255,192,203), 2)
cv2.imshow("Contours on image", result)
cv2.waitKey(0)

# we apply erosions to reduce the size of foreground objects
maskedJar = threshedJar.copy()
maskedJar = cv2.erode(maskedJar, None, iterations=7)
cv2.imshow("Eroded MaskedJar", maskedJar)
cv2.waitKey(0)

# similarly, dilations can increase the size of the ground objects
maskedJar = threshedJar.copy()
maskedJar = cv2.dilate(maskedJar, None, iterations=7)
cv2.imshow("Dilated Masked Jar", maskedJar)
cv2.waitKey(0)

# a typical operation we may want to apply is to take our mask and
# apply a bitwise AND to our input image, keeping only the masked
# regions
mask = threshedJar.copy()
result = cv2.bitwise_and(jarImage, jarImage, mask = mask)
cv2.imshow("Result of bitwise AND", result)
cv2.waitKey(0)
```

```python
## Computing Histogram for each channel of the image
from __future__ import print_function
from matplotlib import pyplot as plt
import numpy as np
import argparse
import cv2

## arparse was commented out because it can only be run on the terminal. to run on terminal, uncomment the next 4 lines
## and comment out the code on line 15
# ap = argparse.ArgumentParser()
# ap.add_argument("-i", "--skyscrapper", required = True, help = "Path to the image")
# args = vars(ap.parse_args())
# skyscrapper = cv2.imread(args["skyscrapper"])

skyscrapper = cv2.imread("skyscrapper.png")
cv2.imshow("Original skyscrapper Image", skyscrapper)

#diplay color_histogram 
chans = cv2.split(skyscrapper)
colors = ("b", "g", "r")
plt.figure()
plt.title("’Flattened’ Color Histogram for Skyscrapper")
plt.xlabel("Bins")
plt.ylabel("# of Pixels")

for (chan, color) in zip(chans, colors):
    hist = cv2.calcHist([chan], [0], None, [256], [0, 256])
    plt.plot(hist, color = color)
    plt.xlim([0, 256])

## Code to automate the process of building a 2D histogram
skyscfig = plt.figure()

ax = skyscfig.add_subplot(131)
hist = cv2.calcHist([chans[1], chans[0]], [0, 1], None, [32, 32], [0, 256, 0, 256])
p = ax.imshow(hist, interpolation = "nearest")
ax.set_title("2D Color Histogram for G and B")
plt.colorbar(p)

ax = skyscfig.add_subplot(132)
hist = cv2.calcHist([chans[1], chans[2]], [0, 1], None, [32, 32], [0, 256, 0, 256])
p = ax.imshow(hist, interpolation = "nearest")
ax.set_title("2D Color Histogram for G and R")
plt.colorbar(p)

ax = skyscfig.add_subplot(133)
hist = cv2.calcHist([chans[0], chans[2]], [0, 1], None, [32, 32], [0, 256, 0, 256])
p = ax.imshow(hist, interpolation = "nearest")
ax.set_title("2D Color Histogram for B and R")
plt.colorbar(p)

print("This is a 2D histogram shape: {}, with {} values".format(hist.shape, hist.flatten().shape[0]))


## Plotting and presenting a 3D histogram
hist = cv2.calcHist([skyscrapper], [0, 1, 2], None, [8, 8, 8], [0, 256, 0, 256, 0, 256])
print("3D histogram shape: {}, with {} values".format(
hist.shape, hist.flatten().shape[0]))

plt.show()
```

### Histogram and Mask

```python
## Function to define our histogram

from matplotlib import pyplot as plt
import numpy as np
import argparse
import cv2

def plot_histogram(skyscrapper, title, mask = None):
    chans = cv2.split(skyscrapper)
    colors = ("b", "g", "r")
    plt.figure()
    plt.title(title)
    plt.xlabel("Bins")
    plt.ylabel("# of Pixels")

    for (chan, color) in zip(chans, colors):
        hist = cv2.calcHist([chan], [0], mask, [256], [0, 256])
    plt.plot(hist, color = color)
    plt.xlim([0, 256])

## arparse was commented out because it can only be run on the terminal. to run on terminal, uncomment the next 4 lines
## and comment out the code on line 28
# ap = argparse.ArgumentParser()
# ap.add_argument("-i", "--skyscrapper", required = True, help = "Path to the image")
# args = vars(ap.parse_args())
# skyscrapper = cv2.imread(args["skyscrapper"])

skyscrapper = cv2.imread("skyscrapper.png")
cv2.imshow("Original Skyscrapper Image", skyscrapper)
plot_histogram(skyscrapper, "Histogram for Original Skyscrapper Image")

## Histogram for masked region
mask = np.zeros(skyscrapper.shape[:2], dtype = "uint8")
cv2.rectangle(mask, (15, 15), (130, 100), 255, -1)
cv2.imshow("Mask skyscrapper", mask)

maskedscrapper = cv2.bitwise_and(skyscrapper, skyscrapper, mask = mask)
cv2.imshow("Applying the Mask", maskedscrapper)

## Plot histogram for Masked Image
plot_histogram(skyscrapper, "Histogram Presentation for Masked Image", mask = mask)
```
